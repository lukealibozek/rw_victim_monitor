```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Send a GET request to the webpage
url = "https://malpedia.caad.fkie.fraunhofer.de/families"
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")

# Find the table containing the desired data
table = soup.find("table")

# # Extract table headers
headers = [th.text.strip() for th in table.find_all("th")]
# headers.insert(0,'index')
# headers.insert(0,'index')
print(headers)
# # Create an empty list to store the data rows
rows = []

# # Extract data from each row
for row in table.find_all("tr"):
    data_row = [td.text.strip() for td in row.find_all("td")]
    # print(data_row)
    if data_row:
        rows.append(data_row)

# # Create a DataFrame from the extracted data
df = pd.DataFrame(rows)
df = df.drop(df.columns[1],axis=1)
df = df.drop(df.columns[1],axis=1)
df.columns = headers
# # Filter and select specific columns
# selected_columns = ["common name", "alt_names", "name", "actors", "entry_updated", "status", "context", "yara"]
# df_selected = df[selected_columns]

# # Print the resulting DataFrame
df
```
